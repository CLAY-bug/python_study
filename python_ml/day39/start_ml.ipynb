```python
from sklearn.datasets import load_iris,fetch_20newsgroups,load_boston,\
    load_digits,load_diabetes
from sklearn.model_selection import train_test_split, GridSearchCV
```

## 加载数据集

### 加载较小的数据集

```python
# 加载鸢尾花数据集
li = load_iris()
print(type(li.data))
print(li.data) # 特征值
print('-' * 30)
print(li.data.shape) # (150, 4) 表示150行数据，4个特征
print('-' * 30)
print(li.target) #
print('-' * 30)
print(li.feature_names) # 特征名，4个特征
print('-' * 30)
print(li.target_names) # 标签名，3个品种
print('-' * 30)
print(li.DESCR)  # 数据描述
```

```python
# 划分鸢尾花的训练集和测试集,注意顺序
# 默认shuffles是True表示是乱序的
# random_state=1是为了却确保每次随机策略是都是一致的
x_train,x_test,y_train,y_test = train_test_split(
    li.data,li.target,test_size=0.25, random_state=1)
print(f'训练集特征值和目标值：{x_train}\n----\n{y_train}')
print(f'训练集特征值shape:{x_train.shape}')
print(f'测试集特征值和目标值：{x_test}\n----\n{y_test}')
print(f'测试集特征值shape:{x_test.shape}')
```

```python
# 加载数字数据集
dig = load_digits()
print(type(dig))  # sklearn.utils._bunch.Bunch 字典格式
print(dig.DESCR)
```

```python
print(dig.data.shape)
print(dig.feature_names)
print('-' * 30)
print(dig.target.shape)
print(dig.target_names)
```

### 下载较大的数据集

```python
# 需要指定下载本地目录
news = fetch_20newsgroups(subset='all', data_home='data')
```

```python
print(news.DESCR)
print('-' * 30)
```

```python
# 数据是一维列表，所有没有shape来查看形状
print(len(news.data)) # 数据总长度
print(len(news.target))
print('-'*30)
# print(news.data[2]) # 查看第二条新闻
print('-'*30)
print(news.target_names)
```
