```python
import matplotlib.pylab as plt
```

```python
# 传入x和y轴元素
plt.plot([1, 0, 9], [4, 5, 6])
plt.show()
```

```python
x = range(1, 8)
y = [17, 17, 18, 15, 11, 11, 13]
plt.plot(x, y)
plt.show()
```

```python
# alpha表示透明度，linestyle表示线条样式，linewidth表示线条宽度
plt.plot(x, y, color='red', alpha=0.5, linestyle='--', linewidth=5)
plt.show()
```

```python
# 折线图
plt.plot(x, y, marker='o')
plt.show()
```

```python
import random

x = range(2, 26, 2)
y = [random.randint(15, 30) for i in x]
# figsize指定宽和高，dpi指定图像的分辨率,默认为80
plt.figure(figsize=(20, 8), dpi=80)
plt.plot(x, y)
# plt.show() #show会释放图片的资源，所以show之后再保存只能保存空图
plt.savefig('./test1.jpg')
```

```python
x = range(2, 26, 2)
y = [random.randint(15, 30) for i in x]
# x轴的刻度标签
x_ticks_label = ["{}:00".format(i) for i in x]
plt.xticks(x, x_ticks_label, rotation=45)
# y轴的刻度标签
y_ticks_label = ["{}·C".format(i) for i in range(min(y), max(y) + 1)]
plt.yticks(range(min(y), max(y) + 1), y_ticks_label)
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
```

```python
x = range(0, 120)
y = [random.randint(10, 30) for i in range(120)]
plt.figure(figsize=(20, 8), dpi=80)
plt.plot(x, y)
# matlplotlib需要自己下载字体导入以支持中文
plt.xlabel('时间', fontproperties='simHei', fontsize=20)
plt.ylabel('次数', fontproperties='simHei')
plt.title('每分钟跳动次数', color='red')
plt.show()
```
