```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

```python
my_font = font_manager.FontProperties(fname='/System/Library/Fonts/PingFang.ttc', size=18)
```

```python
y1 = [1, 0, 1, 1, 2, 4, 3, 4, 4, 5, 6, 5, 4, 3, 3, 1, 1, 1, 1, 1]
y2 = [1, 0, 3, 1, 3, 4, 6, 3, 5, 5, 3, 1, 3, 4, 5, 7, 1, 3, 4, 1]
x = range(11, 31)
plt.figure(figsize=(20, 8), dpi=80)
plt.plot(x, y1, color='red', label='自己')  # 第一条折线
plt.plot(x, y2, color='blue', label='朋友')  # 第二条折线
xtick_labels = ['{}岁'.format(i) for i in x]
plt.title('恋爱史', fontproperties=my_font)
plt.xticks(x, xtick_labels, fontproperties=my_font, rotation=45)  # 设置x轴的刻度和显示中文字体
plt.ylabel('恋爱个数', fontproperties=my_font)
plt.grid(alpha=0.5)  # 设置透明度
plt.legend(prop=my_font, loc='upper left')  # 设置图例，放在左上角
plt.show()
```

```python
import numpy as np
```

```python
x = np.arange(1, 100)
fig, axis = plt.subplots(2, 2)
ax1 = axis[0, 0]
ax2 = axis[0, 1]
ax3 = axis[1, 0]
ax4 = axis[1, 1]

plt.figure(figsize=(10, 10), dpi=80)
ax1.grid(alpha=0.4)
ax1.plot(x, x)
ax2.plot(x, -x)
ax3.plot(x, x ** 2)
ax4.plot(x, np.log(x))
plt.show()
```

散点图

```python
y = [11, 17, 16, 11, 12, 11, 12, 6, 6, 7, 8, 9, 12, 15, 14, 17, 18, 21, 16, 17, 20, 14,
     15, 15, 15, 19, 21, 22, 22, 22, 23]
x = range(1, 32)
plt.figure(figsize=(20, 8), dpi=80)
plt.scatter(x, y, label='3月')
xtick_labels = ['3月{}日'.format(i) for i in x]
plt.xticks(x[::2], xtick_labels[::2], fontproperties=my_font, rotation=45)  # 设置x轴刻度，每2日显示一次
plt.xlabel('日期', fontproperties=my_font)
plt.ylabel('温度', fontproperties=my_font)
plt.legend(prop=my_font)
plt.show()
```

绘制条形图

```python
a = ['流浪地球', '疯狂的外星人', '飞驰人生', '大黄蜂', '熊出没·原始时代', '新喜剧之王']
b = [38.13, 19.85, 14.89, 11.36, 6.47, 5.93]
plt.figure(figsize=(20, 8), dpi=80)
rects = plt.bar(range(len(a)), b, width=0.4, color='blue')  # 返回的是条形图的对象，个数为a的长度
for rect in rects:  # 依次对每个条形进行设置标注
    height = rect.get_height()  #获取条形的高度，并在柱上显示
    plt.text(rect.get_x() + rect.get_width() / 2, height + 0.5, str(height), ha='center')
plt.show()
```

直方图

```python
with open('/Users/clay/python_code6/数据分析/day32/film_time', encoding='utf8') as f:
    time = f.read()
time = time.split(',')
for i in range(len(time)):
    time[i] = int(time[i])
```

```python
plt.figure(figsize=(20, 8), dpi=80)
distance = 2  # 设置距组
group_num = int((max(time) - min(time)) / distance)
plt.hist(time, bins=group_num)
plt.xticks(range(min(time), max(time))[::2])  #设置x轴的刻度信息，从最短时长到最长时长，以2个单位递增
plt.grid(alpha=0.5)
plt.xlabel('电影时长', fontproperties=my_font)  # 设置x和y轴描述信息
plt.ylabel('电影数量', fontproperties=my_font)
plt.show()
```

饼状图

```python
label_list = ["第一部分", "第二部分", "第三部分"]  #各部分的名称
size = [34, 46, 20]  # 各部分所占面积大小
color_list = ['red', 'purple', 'blue']
explode_list = [0, 0.06, 0]
patches, l_text, p_text = plt.pie(size, explode=explode_list, colors=color_list, labels=label_list,
                                  labeldistance=1.1, autopct='%1.1f%%')
for t in l_text:
    t.set_fontproperties(my_font)
plt.legend(prop=my_font, loc='upper right')
plt.axis('equal')
plt.show()
```
